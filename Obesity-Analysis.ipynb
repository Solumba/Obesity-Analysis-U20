```python
#We'll import our libraries here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure style
sns.set_style("whitegrid")

# Define categorical variables to analyze
categorical_features = ["Gender", "family_history_with_overweight", "FAVC", "CAEC", "SMOKE", "CALC", "MTRANS"]

# Set up subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
axes = axes.flatten()

# Create bar plots for each categorical feature
for i, feature in enumerate(categorical_features[:6]):  # First 6 features for plotting
    sns.countplot(x=feature, hue="NObeyesdad", data=df, ax=axes[i])
    axes[i].set_title(f"{feature} vs. Obesity Category")
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("Count")
    axes[i].legend(title="Obesity Category", loc="upper right")

# Adjust layout
plt.tight_layout()
plt.show()
```
